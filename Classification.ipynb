## Multiclass classification

```python
import tensorflow as tf
from tensorflow.keras.datasets import fashion_mnist

(train_data,train_labels),(test_data,test_labels) = fashion_mnist.load_data()
train_data = train_data/255.0
test_data = test_data/255.0
```

```python
tf.random.set_seed(42)

model_m = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28,28)),
    tf.keras.layers.Dense(4,activation='relu'),
    tf.keras.layers.Dense(4,activation='relu'),
    tf.keras.layers.Dense(10,activation=tf.keras.activations.softmax)
])

model_m.compile(loss=tf.keras.losses.SparseCategoricalCrossentropy(),
                optimizer=tf.keras.optimizers.Adam(),
                metrics=["accuracy"])

history = model_m.fit(train_data,train_labels,epochs=10,validation_data=(test_data,test_labels),verbose=1)
model_m.evaluate(test_data,test_labels)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

pd.DataFrame(history.history).plot()
plt.show()
```
